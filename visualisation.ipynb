# BCI IV 2a Epochs

```python
import os, mne
import matplotlib.pyplot as plt
```

```python
directory = "./preprocessed_data/BCI_IV_2a"

s1_epochs = os.path.join(directory, "S01", "PA01T-epo.fif")
epochs = mne.read_epochs(s1_epochs)
epochs
```

## Evoked left hand

```python
evoked_left = epochs['left_hand'].average()

evoked_left.plot()
plt.show()
```

## Evoked right hand

```python
evoked_right = epochs['right_hand'].average()

evoked_right.plot()
plt.show()
```

## Compare evoked

```python
mne.viz.plot_compare_evokeds({"left": evoked_left, "right": evoked_right})
```
